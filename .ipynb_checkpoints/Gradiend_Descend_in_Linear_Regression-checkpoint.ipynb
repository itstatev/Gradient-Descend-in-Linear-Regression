```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
x = 2 * np.random.randn(m,1)
y = 4 + 2 * x + np.random.randn(m, 1) #generating random errors
```

```python
plt.scatter(x,y)
```

```python
x_b = np.c_[np.ones((m,1)), x] #Adding a column with ones as the first column
```

```python
alpha = 0.1 #learning rate
```

```python
iterations = 100
```

```python
theta = np.random.randn(2,1) 
```

```python
for iteration in range(iterations):
    gradients = (2 / m) * x_b.T.dot(x_b.dot(theta) - y)
    theta = theta - alpha * gradients
    pred = theta[0] + theta[1] * x
```

```python
plt.scatter(x, y)
plt.plot([min(x), max(x)], [min(pred), max(pred)], color = 'red')
```
